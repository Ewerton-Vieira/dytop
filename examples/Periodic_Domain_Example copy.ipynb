```python
import CMGDB
import matplotlib.pyplot as plt
import numpy as np
```

# A silly example of a torus map

```python
# Define the torus map
# f(x, y) = (2 x + 3 y, 0.5 x + 0.3 y) mod 1
def f(x):
    return [(2.0 * x[0] + 3.0 * x[1]) % 1, (0.5 * x[0] + 0.3 * x[1]) % 1]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

```python
subdiv_min = 20
subdiv_max = 30
lower_bounds = [0.0, 0.0]
upper_bounds = [1.0, 1.0]
phase_periodic = [True, True]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, phase_periodic, F)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph)
```

### Periodic example

```python
# Define the torus map
# f(x, y) = (2 x + 3 y, 0.5 x + 0.3 y) mod 1
def f(x):
    x[0] = ( (x[0] - 1/2) ** 3 + 1/2 ) % 1

    x[1] = np.cbrt((x[1] - 1/2)) + 1/2
    
    if x[1] >= 1:
        x[1] = 1
        
    if x[1]<= 0:
        x[1] = 0
    
    return [x[0], x[1]]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

```python
subdiv_min = 20
subdiv_max = 30
lower_bounds = [0.0, 0.0]
upper_bounds = [1.0, 1.0]
phase_periodic = [True, True]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, phase_periodic, F)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph)
```

### Periodic example

```python
# Define the torus map
# f(x, y) = (2 x + 3 y, 0.5 x + 0.3 y) mod 1
def f(x):
    x[0] = (x[0] - 1) ** 3 + 1

    x[1] = np.cbrt((x[1] - 1)) + 1
    
    
    return [x[0], x[1]]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

### Periodic example

```python
# Define the torus map
# f(x, y) = (2 x + 3 y, 0.5 x + 0.3 y) mod 1
def f(x):
    x[0] = ( ((x[0] - 1)) ** 3 + 1 ) 

#     x[0] = x[0]

#     x[1] = ( ((x[1] - 1)) ** 3 + 1 ) % 2
    
    x[1] = (np.cbrt((x[1] - 1)) + 1) * x[0]
    
    
    return [x[0], x[1]]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```
